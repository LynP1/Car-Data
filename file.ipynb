```python
import pandas as pd
data = pd.read_csv('fifa_2023.csv')
```

```python
data.corr(numeric_only=True)
```

```python
data.plot(kind='scatter',x='rating',y='passing')
```

<h1>Shooting has a strong correlation with Dribbling</h1>

```python
data.plot(kind='scatter',x='shooting',y='dribbling')
```

<h1>Defense has a slight correlation with Physicality</h1>

```python
data.plot(kind='scatter',x='defense',y='physicality')
```
